```python
import numpy as np
import pandas as pd
temp_data = pd.read_csv('/content/daily-min-temperatures.csv')
temp_data.head(15)
```

```python
#look at the information in our dataset
temp_data.info()
```

```python
# Convert Date to a datetime column
temp_data['Date'] = pd.to_datetime(temp_data['Date'], format='%Y-%m-%d')
```

```python
temp_data.info()
```

```python
temp_data.head(15)
```

```python
temp_data = temp_data.set_index('Date')
temp_monthly = temp_data.resample('MS')
month_mean = temp_monthly.mean()
display(month_mean)
```

```python
month_mean.head(15)
```

```python
temp_bidaily = temp_data.resample('12H').asfreq()
temp_bidaily.head()
```

```python
temp_bidaily_fill = temp_bidaily.ffill()
temp_bidaily_fill.head()
```

```python
 #slice our dataset to only retrieve data points that come after the year 1985
 temp_1985_onwards = temp_data['1985':]
print(temp_1985_onwards.head())
print(temp_1985_onwards.tail())
```
